```python
import pandas as pd
```

```python
df =pd.read_excel(r'C:\Users\paras\Desktop\task\college_cutoffs.xlsx')
```

```python
new_df = pd.DataFrame()
```

```python
j=0
Branch_name =[]
Rank =[]
college_name =[]
```

```python

for i in df['D']:
    j = j+1
    if ('Savitribai Phule Pune University' in str(i)):
        
       
        if('Civil Engineering' in df['D'][j-2]):
                   
    
            if(df['Unnamed: 1'][j+1]=='GOPENH' or df['Unnamed: 1'][j+1]=='GOPENS'):
                Branch_name.append(df['D'][j-2])
                college_name.append(df['D'][j-3])
                Rank.append(df['Unnamed: 1'][j+2])
       
```

```python
new_df['Branch_name']=Branch_name
new_df['Rank']=Rank
new_df['college_name']=college_name
```
